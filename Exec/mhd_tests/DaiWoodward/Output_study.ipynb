```python
from matplotlib import pyplot as plt

import yt
from yt import derived_field
import numpy as np
```

```python
ds = yt.load("plt02419/")
```

```python
ray = ds.ortho_ray(0,(0,0))
ray_sort = np.argsort(ray["x"])

x = np.array(ray['x'][ray_sort])


plt.plot(x,np.array(ray['density'][ray_sort]))
plt.show()
```

```python
plt.plot(x, np.array(ray['z_velocity'][ray_sort]))
plt.show()
```
